```python
bike_rent2 = pd.concat([sum_rent1,sum_rent2, sum_rent3, sum_rent4, sum_rent5, sum_rent6], axis =1)
bike_rent2
```

```python
bike_rent2.columns = ['1월', '2월', '3월', '4월', '5월', '6월']
bike_rent2['총 이용량'] = bike_rent2['1월'] +  bike_rent2['2월'] +  bike_rent2['3월'] +  bike_rent2['4월'] +  bike_rent2['5월'] +  bike_rent2['6월']
```

```python
bike_rentsum = pd.DataFrame(bike_rent2)
bike_rentsum
```

```python
bike_allrent = bike_rentsum.drop(['1월', '2월', '3월', '4월', '5월', '6월'], axis=1)
bike_allrent
```

```python
bike_allrent.reset_index(drop =False, inplace = True)
bike_allrent
```

```python
plt.figure(figsize=(25,10))
plt.xticks(rotation = 45)
plt.rc('font', size=25)   
plt.ticklabel_format(axis='y', style='plain')
plt.title('서울시 공공자전거 총 이용량')
sns.barplot(x = '자치구', y = '총 이용량', data = bike_allrent)
plt.show()
```
